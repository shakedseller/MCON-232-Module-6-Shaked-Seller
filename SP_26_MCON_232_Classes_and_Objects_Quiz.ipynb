## Short reminders before you begin
A **class** is a blueprint.  
An **object** is one instance created from that blueprint.  
The `__init__` method builds the starting state of each object.  
An **instance attribute** belongs to one object, usually written with `self.attribute_name`.  
A **class attribute** belongs to the class itself and is shared unless shadowed by an instance.  
A **method** is a function inside a class.  
When an object is passed into a method, the method receives a reference to that object.  
If the method **mutates the object's attributes**, the caller sees those changes after the method ends.  
If the method **reassigns the parameter variable** to a different object, that reassignment only changes the local variable inside the method.

Keep these ideas in mind while working:

- Instance methods usually begin with `self`
- Access instance data with `self.attribute`
- Access class data with `ClassName.attribute`

- Passing an object into another object's method allows one object to use the other object's data

### Question B3 — Passing one object into another object's method

Before running the code, answer:

```python

class Student:
    def __init__(self, name, grade):
        self.name = name
        self.grade = grade

class Reporter:
    def show_student(self, student_obj):
        print(student_obj.name, student_obj.grade)

s1 = Student("Leah", 95)
r = Reporter()
r.show_student(s1)
```

### Question B4 — Mutation inside a method

Before running the code, answer:
- What is printed at the end?
- Why does that change remain after the method call?

```python

class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance

class BankTeller:
    def deposit_bonus(self, account, amount):
        account.balance += amount

acct = BankAccount("Mina", 100)
teller = BankTeller()
teller.deposit_bonus(acct, 25)

print(acct.balance)
```

**Your answer:**
- 125 is printed at the end, it remains after the method call because the object is passed as a reference

### Question B5 — Reassigning the parameter inside a method

Before running the code, answer:
- What is printed?
- Why does the original object stay the same?

```python

class Box:
    def __init__(self, value):
        self.value = value

class Changer:
    def replace_box(self, box_obj):
        box_obj = Box(999)
        box_obj.value += 1

b = Box(10)
c = Changer()
c.replace_box(b)

print(b.value)
```

**Your answer:**
- 10 is printed, the parameter inside replace_box() is being replaced, so it doesn't change the original object.

## Quick Vocabulary

Before we write code, keep these words in mind:

- **class** → a blueprint
- **object / instance** → one thing made from the blueprint
- **attribute** → data stored inside the object
- **method** → a function that belongs to the object
- **instantiate** → create an object from a class

Example idea:

- `Dog` → class
- `d1 = Dog("Fido", 3)` → object instantiation
- `name` and `age` → attributes
- `bark()` → method

## Example 1 — Reading a Class

Study this example first. Then run it.

````python

class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def bark(self):
        return f"{self.name} says woof!"

    def birthday(self):
        self.age += 1


d1 = Dog("Fido", 3)
d2 = Dog("Luna", 5)

print(d1.bark())
print(d2.bark())

d1.birthday()
print(d1.age)
````

### Think about it
Answer these in your own words:

1. What is the class name?
- The class name is Dog
2. What are the attributes?
- The attributes are self.name and self.age
3. What is the method `birthday()` changing?
- It is changing self.age
4. Why do `d1` and `d2` keep different data?
- d1 and d2 are two separate instances of Dog

## Exercise 1 — Build a `Book` Class

Create a class called `Book`.

### Attributes
Each `Book` object should store:

### Methods
Add a method called `describe()` that returns a string in this form:

`"<title> by <author>, <pages> pages"`

### Example
`Book("Matilda", "Roald Dahl", 240).describe()`

should return:

`"Matilda by Roald Dahl, 240 pages"`

````python

class Book:
    # TODO: write __init__
    def __init__(self, title, author, pages):
      self.title = title
      self.author = author
      self.pages = pages

    # TODO: write describe
    def describe(self):
      return f"{self.title} by {self.author}, {self.pages} pages"
````

## Exercise 2 — Build a `Student` Class

Create a class called `Student`.

### Attributes
- `name`
- `grade`

### Methods
1. `introduce()`  
   Returns: `"Hi, my name is <name>."`

2. `update_grade(new_grade)`  
   Changes the student's grade to `new_grade`

````python

class Student:
    # TODO: write __init__
    def __init__(self, name, grade):
      self.name = name
      self.grade = grade

    # TODO: write introduce
    def introduce(self):
      return f"Hi, my name is {self.name}."

    # TODO: write update_grade
    def update_grade(self, new_grade):
      self.grade = new_grade
````

## Exercise 4 — Build a `Rectangle` Class

Create a class called `Rectangle`.

### Attributes
- `width`
- `height`

### Methods
1. `area()` → returns width * height
2. `perimeter()` → returns 2 * (width + height)

````python

class Rectangle:
    # TODO: write __init__
    def __init__(self, width, height):
      self.width = width
      self.height = height
    # TODO: write area
    def area(self):
      return self.width * self.height
    # TODO: write perimeter
    def perimeter(self):
      return 2 * (self.width + self.height)
````

````python

r1 = Rectangle(5, 3)
assert_equal(r1.width, 5)
assert_equal(r1.height, 3)
assert_equal(r1.area(), 15, "area() is incorrect.")
assert_equal(r1.perimeter(), 16, "perimeter() is incorrect.")
print("✅ Exercise 4 passed")
````

## Exercise 5 — Build a `BankAccount` Class

Create a class called `BankAccount`.

### Attributes
- `owner`
- `balance`

### Methods
1. `deposit(amount)`  
   Adds `amount` to the balance

2. `withdraw(amount)`  
   If there is enough money, subtract `amount`  
   Otherwise, do **not** change the balance

3. `display_balance()`  
   Returns: `"<owner> has $<balance>"`

````python

class BankAccount:
    # TODO: write __init__
    def __init__(self, owner, balance):
      self.owner = owner
      self.balance = balance
    # TODO: write deposit
    def deposit(self, amount):
      self.balance += amount
    # TODO: write withdraw
    def withdraw(self, amount):
      if amount <= self.balance:
        self.balance -= amount
    # TODO: write display_balance
    def display_balance(self):
      return f"{self.owner} has ${self.balance}"
````

````python

acct = BankAccount("Rivka", 100)
assert_equal(acct.owner, "Rivka")
assert_equal(acct.balance, 100)

acct.deposit(40)
assert_equal(acct.balance, 140, "deposit() should add to the balance.")

acct.withdraw(25)
assert_equal(acct.balance, 115, "withdraw() should subtract when there is enough money.")

acct.withdraw(1000)
assert_equal(acct.balance, 115, "withdraw() should not allow the balance to go below zero.")

assert_equal(acct.display_balance(), "Rivka has $115")
print("✅ Exercise 5 passed")
````

## Exercise 7 — Flower Warm-Up (No Turtle Yet)

Before we do graphics, let us model a flower as an object.

Create a class called `Flower`.

### Attributes
- `name`
- `petals`
- `height`

### Methods
1. `describe()`  
   Returns: `"<name> has <petals> petals and is <height> cm tall."`

2. `grow(cm)`  
   Adds `cm` to the flower's height

````python

class Flower:
    # TODO: write __init__
    def __init__(self, name, petals, height):
      self.name = name
      self.petals = petals
      self.height = height
    # TODO: write describe
    def describe(self):
      return f"{self.name} has {self.petals} petals and is {self.height} cm tall."
    # TODO: write grow
    def grow(self, cm):
      self.height += cm
````

````python

f1 = Flower("Rose", 12, 15)
assert_equal(f1.name, "Rose")
assert_equal(f1.petals, 12)
assert_equal(f1.height, 15)
assert_equal(f1.describe(), "Rose has 12 petals and is 15 cm tall.")

f1.grow(4)
assert_equal(f1.height, 19, "grow(cm) should increase the height.")
print("✅ Exercise 7 passed")
````

## Big Reminders
A **class** is a blueprint for making objects.  
An **object** is one instance of that class.  
The constructor `__init__` gives each object its starting values.  
Instance attributes usually belong to one object.  
Class attributes belong to the class and are shared unless shadowed.  
Methods let objects do work.  
One object can receive another object as a parameter and use that object's data.  
If a method mutates an object, that mutation is still there when the method returns.  
If a method reassigns the parameter variable to a different object, the original caller's object does not get replaced.

- constructor parameter:
  - `username`

### Reminder
A class variable can track information shared by all objects.

### Requirements  
Counting must take place within a method of the class

````python

# Write your answer here
class User:
  total_users = 0

  def __init__(self, username : str) -> None:
    self.username = username
    User.increase_users()

  @classmethod
  def increase_users(cls):
    cls.total_users += 1

user1 = User("tsukkomicoder")
user2 = User("almondblossom")
user3 = User("windexcube")

print(User.total_users)
````

### `Gardener`
Write two methods:

1. `water(plant_obj, inches)`

   - reassign `plant_obj` to a new `Plant("Replacement", 100)`

### After writing the classes

### Reflection

````python

# Write your answer here
class Plant:
  def __init__(self, name : str, height : int) -> None:
    self.name = name
    self.height = height

class Gardener:
  def water(self, plant_obj, inches):
    plant_obj.height += inches

  def replace_local(self, plant_obj):
    plant_obj = Plant("Replacement", 100)

plant = Plant("Cosmos", 50)
g = Gardener()

g.water(plant, 20)
print(plant.name, plant.height)

g.replace_local(plant)
print(plant.name, plant.height)

# Water does change the original object because it is being passed by reference,
# therefore modifying the original object

# Replace_local does not change the object because it reassigns the parameter,
# so it isn't editing the original object anymore
````

## Task 9 — A `BankAccount` class

Create a class named `BankAccount`.

### Requirements
- constructor parameters:
  - `owner`
  - `balance`
- method `deposit(amount)`
  - add to the balance
- method `withdraw(amount)`
  - if enough money exists, subtract and return `True`
  - otherwise do not change the balance and return `False`
- method `display()`
  - return a string like:
    - `"Ari has $120."`

Create one account and test all methods.

### Reminder
This is classic object-oriented design: the object stores state and methods manage that state.

````python

# Write your answer here
class BankAccount:
  def __init__(self, owner : str, balance : int) -> None:
    self.owner = owner
    self.balance = balance

  def deposit(self, amount : int):
    self.balance += amount

  def withdraw(self, amount : int) -> bool:
    if self.balance >= amount:
      self.balance -= amount
      return True
    else:
      return False

  def display(self):
    return f"{self.owner} has ${self.balance}"

acc = BankAccount("Shaked", 1000)
acc.deposit(500)
print(acc.withdraw(50))
print(acc.withdraw(999999999))

print(acc.display())
````

## Task 10 — A `Classroom` and `Student`

Create two classes: `Student` and `Classroom`.

### `Student`
- constructor parameters:
  - `name`
  - `grade`

### `Classroom`
- constructor should create an empty list named `students`
- method `add_student(student_obj)`
  - add the object to the list
- method `highest_grade()`
  - return the highest grade among the students
  - if there are no students, return `None`

### Test
- create at least 3 students
- add them to a classroom
- print the highest grade

### Reminder
This is good practice with objects containing collections of other objects.

````python

# Write your answer here
class Student:
  def __init__(self, name : str, grade : int) -> None:
    self.name = name
    self.grade = grade

class Classroom:
  def __init__(self):
    self.students = []

  def add_student(self, student_obj):
    self.students.append(student_obj)

  def highest_grade(self):
    if len(self.students) > 0:
      highest = 0
      for s in self.students:
        if s.grade > highest:
          highest = s.grade
      return highest

    else:
      return None


s1 = Student("S", 99)
s2 = Student("D", 98)
s3 = Student("C", 97)

c = Classroom()

c.add_student(s1)
c.add_student(s2)
c.add_student(s3)

print(c.highest_grade())
````

## Task 11 — Design + object interaction challenge

Create two classes: `Player` and `InventoryManager`.

### `Player`
- constructor parameters:
  - `name`
  - `items`  (this will be a list of strings)

### `InventoryManager`
Add two methods:

1. `add_item(player_obj, item)`
   - append the item to the player's `items` list

2. `trade_first_item(player_a, player_b)`
   - if **both** players have at least one item:
     - swap their first items
     - return `True`
   - otherwise return `False`

### Test
Create two players with different item lists.
Demonstrate both methods.

### Why this is a thinking question
You must manage state stored inside objects, including a mutable list attribute.
Be careful to mutate the correct object data.

````python

# Write your answer here
class Player:
  def __init__(self, items : list) -> None:
    self.items = items

class InventoryManager:
  def add_item(self, player_obj, item : str):
    player_obj.items.append(item)

  def trade_first_item(self, player_a, player_b) -> bool:
    if len(player_a.items) >= 1 and len(player_b.items) >= 1:
      player_a_item = player_a.items[0]
      player_b_item = player_b.items[0]

      player_a.items[0] = player_b_item
      player_b.items[0] = player_a_item

      return True

    else:
      return False

p1 = Player(["Hylian Shrooms", "Staminoka Bass", "Glowing Cave Fish"])
p2 = Player(["Rock Salt", "Tabantha Wheat", "Hylian Herb"])

im = InventoryManager()

print(im.trade_first_item(p1, p2))
print(p1.items)
print(p2.items)
````

## Task 12 — LeetCode-style object design challenge

Create a class named `ScoreTracker`.

### Requirements
- constructor creates an empty dictionary to store player scores
- method `add_score(player_name, points)`
  - if the player is new, create the entry
  - otherwise add to the existing score
- method `top_player()`
  - return the name of the player with the highest score
  - if there are no players, return `None`
- method `top_score()`
  - return the highest score
  - if there are no players, return `None`

### Test
Add several scores for several players and show that repeated calls accumulate correctly.

### Reflection
In a comment, explain why this is a good example of a class:
- what state is being stored?
- what behaviors are being grouped with that state?

````python

# Write your answer here
class ScoreTracker:
  def __init__(self):
    self.scores = {}

  def add_score(self, player_name : str, points :int) -> None:
    if player_name not in self.scores:
      self.scores[player_name] = points

    else:
      self.scores[player_name] += points

  def top_player(self):
    if len(self.scores) == 0:
      return None

    highest = 0
    highest_name = ""
    for k, v in self.scores.items():
      if v > highest:
        highest = v
        highest_name = k

    return highest_name

  def top_score(self):
    return max(self.scores.values())

st = ScoreTracker()

st.add_score("Player1", 10)
st.add_score("Player2", 20)
st.add_score("Player1", 20)

print(st.scores)
print(st.top_player())
print(st.top_score())

st.add_score("Player2", 50)

print(st.scores)
print(st.top_player())
print(st.top_score())

# The state being stored is the dictionary that keeps track of the players
# and their scores

# The behaviors that are grouped with that state are the member functions
# of the class
````
